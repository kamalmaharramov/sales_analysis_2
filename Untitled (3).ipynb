```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
sales_data=pd.read_csv("Diwali Sales Data.csv",encoding='latin1')
sales_data
```

# 1.drop status and unnamed1 columns

```python
sales_data.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
sales_data.shape
```

```python
sales_data.isnull().sum()
```

```python
sales_data[sales_data["Amount"].isnull()]
```

# drop null values

```python
sales_data.dropna(inplace=True)
sales_data
```

```python
corr=sales_data.corr(method="pearson")
from matplotlib.colors import ListedColormap
sns.heatmap(corr,cmap=ListedColormap(['green', 'yellow', 'red',"blue","black"]))
```

# gender

```python
sns.countplot(x="Gender",data=sales_data)
```

# sales by gender

```python
sales_by_gender=sales_data.groupby(["Gender"]).sum()
sales_by_gender
plt.bar(sales_by_gender.index,sales_by_gender["Amount"],color=("blue","red"))
```

```python
sales_by_age=sales_data.groupby(["Age Group"])
sales_by_age
```

```python
plt.plot(sales_by_age.sum().index,sales_by_age.sum()["Amount"],color=("red"),marker="o",mec="black",markersize=15)
plt.grid()
plt.show()
```

# for state

```python
sales_by_state=sales_data.groupby(["State"]).sum()
```

```python
plt.plot(sales_by_state.index,sales_by_state["Amount"],color=("blue"),marker="o",mec="red",markersize=15)
plt.xticks(rotation="vertical")
plt.grid()
plt.show()
```
